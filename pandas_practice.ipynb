```python
#loc사용법
df1 = pd.read_csv('train.csv')
df1.head()
df1.loc[:,:]  #전체 dataframe 출력
df1.loc[1] #df.loc[특정 행의 index]
df1['Name']  #df [특정 column index] = df.loc[:, column index]  
df1.loc[:1,'Name']  # 0~1행에서 column index가 'Name'의 값을 출력한다.
df1.loc[:4,:'Age']   # 내가 원하는 범위의 row, column units 출력한다.
```

```python
df1.loc[df1['Pclass'] == 3]   #Pclass 라는 column에서 3인 값들만 추출한다.
df1.loc[(df1['Pclass']==3 )| (df1['Age']>=30)]   #  ( ) | (  ) 는 or을 뜻함
df1.loc[(df1['Pclass']==3 )& (df1['Age']>=30)]   #  ( ) & (  ) 는 and을 뜻함
```

```python
#iloc 사용법 = integer location : df의 row or column 순서를 나타내는 정수로 특정 값 추출 ->컴퓨터 친화적
df1.iloc[0]   # df의 0번째 행의 값들 추출
df1.iloc[0,3]   #df.iloc[행,열] :  위에서 출력한 행의 특정 column 위치의 값을 출력함
df1.iloc[:5,:5]   # 행 5개, 열 5개 출력 : loc과는 다르게 마지막 쓴 숫자 전까지만 출력됨
df1.iloc[::2,:]   # ::2 는 행을 2 간격으로 출력하는 것
#주의 column간에 무조건 정수만 들어가야 함, loc 처럼 'Pclass' 같은 거 쓰면 안 되고 그 자리의 위치를 정수로 입력해주어야 함

 
```

```python
df4=pd.read_csv('train.csv')
df4.head()
```

```python
df4['Age'].sort_values()  #특정 column의 값을 오름차순(descending)으로 정렬(1,2,3 , a,b,c), NaN값은 맨 마지막에
df4['Age'].sort_values(ascending = False)  #내림차순(ascending)
```
